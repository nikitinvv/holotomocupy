```python
import numpy as np
import tifffile
from holotomocupy.utils import *

prec_path = '/data2/vnikitin/brain/20251115/Y350a_HT_20nm_8dist_rec1234_/newFeb2026/Y350a_HT_20nm_8dist_rec_cm_.vol'
dtype = np.dtype("<f4")

with open(prec_path, "rb") as f:
    buf = f.read(nz * ny * nx * dtype.itemsize)

prec = np.frombuffer(buf, dtype=dtype).reshape(nz, ny, nx)
```

```python
iter=1184
# mrec = tifffile.imread(f'/data2//vnikitin/brain_rec/20251115/Y350a1234/4500_2048_0_0.0_0.003_0.05_0.02_20_1.1_0/rec_obj_real/{iter:04}.tiff')
with h5py.File(f'/data2/vnikitin/alcf/brain/Y350a_dist1234/rec_levels_17p23/checkpoint_{iter}.h5','r') as fid:
    mrec = fid['/obj_re']
    mrec = mrec[mrec.shape[0]//2-nz//2:mrec.shape[0]//2+nz//2,
            mrec.shape[1]//2-ny//2:mrec.shape[1]//2+ny//2,
            mrec.shape[2]//2-nx//2:mrec.shape[2]//2+nx//2]
```

```python
mpl.rcParams['font.size'] = 22  # optional
mpl.rcParams["xtick.labelsize"] = 16
mpl.rcParams["ytick.labelsize"] = 16

from matplotlib.colors import LinearSegmentedColormap

# white -> black colormap
wb = LinearSegmentedColormap.from_list("white_black", ["white", "black"])

def mshow(a, show=False, **args):
    if show:
        if isinstance(a, cp.ndarray):
            a = a.get()
        fig, axs = plt.subplots(1, 1, figsize=(6,6))
        im = axs.imshow(a, cmap=wb,**args)
        cbar = fig.colorbar(im, fraction=0.046, pad=0.1,orientation="horizontal", location="bottom")
        scalebar = ScaleBar(20e-3, "um", length_fraction=0.25, font_properties={
            "family": "serif", "size":20
        },  # For more information, see the cell below                
        location="lower right")
        ticks = np.array(cbar.get_ticks())
        if ticks.size >= 2:
            cbar.set_ticks([ticks[0], ticks[-1]])
            cbar.set_ticklabels([f"{ticks[0]:g}", f"{ticks[-1]:g}"])

        cbar.set_label("Electron density (e/Å$^3$)", labelpad=-22, fontsize=17)

        cbar.ax.tick_params(labelsize=18)
        # axs.axis('off')
        axs.add_artist(scalebar)
        
        # plt.show()
def find_min_max(data):
    """Find min and max values according to histogram"""

    h, e = np.histogram(data[:], 1000)
    stend = np.where(h > np.max(h)*0.01)
    st = stend[0][0]
    end = stend[0][-1]
    mmin = e[st]
    mmax = e[end+1]

    return mmin, mmax

import numpy as np

# constants
r_e = 2.8179403262e-15  # classical electron radius [m]
pi = np.pi

def wavelength_from_keV(E_keV):
    """X-ray wavelength [m] from photon energy [keV]."""
    hc = 1.239841984e-6  # [eV*m]
    return hc / (E_keV * 1e3)

def m3_to_A3(rho_m3):
    """Convert electrons/m^3 to electrons/Å^3."""
    return rho_m3 * 1e-30  # since 1 m^3 = 1e30 Å^3

def m2_to_A2(Ne_m2):
    """Convert electrons/m^2 to electrons/Å^2."""
    return Ne_m2 * 1e-20  # since 1 m^2 = 1e20 Å^2

def rho_e_from_delta_A3(delta, lam_m):
    """rho_e [e/Å^3] from delta."""
    delta = -delta*1e-9*1.2
    rho_m3 = (2*np.pi/(r_e*lam_m**2)) * delta
    return m3_to_A3(rho_m3)
```
