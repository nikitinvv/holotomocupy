# Step 3. Combine multi-source shifts into a single calibrated shift file

Four independent sources of sample displacement are measured separately and then summed to produce
the final per-projection, per-distance shift field used in Steps 4 and 5:

| Source | Variable | Description |
|--------|----------|-------------|
| Encoder readback | `random_shifts` | Sample stage jitter reported by the scan motor (`correct.txt`) |
| RHAPP inter-plane | `rhapp_shifts` | Lateral offset *between* distance planes from Peter's MATLAB alignment |
| Motion correction | `motion_shifts` | Slow drift of the reference plane over the full 360° scan |
| 3-D tomo correction | `correct3d_shifts` | Residual shifts after a preliminary tomographic reconstruction |

All shifts are converted to **object pixels at the highest magnification** (distance 0) before
summation.

```python
import numpy as np
import cupy as cp
import h5py
import matplotlib.pyplot as plt
from holotomocupy.utils import *
```

## Init

```python
ntheta = 1800
ref_dist = 0
ids = np.arange(0, 1800, 1800 / ntheta).astype('int')

path     = '/data2/vnikitin/atomium/20240924/AtomiumS2/'
pfile    = 'AtomiumS2_HT_007nm'
path_out = '/data2/vnikitin/atomium_rec/20240924/AtomiumS2/'
file_out = 'data.h5'

with h5py.File(f'{path_out}/{file_out}') as fid:
    detector_pixelsize      = fid['/exchange/detector_pixelsize'][0]
    focustodetectordistance = fid['/exchange/focusdetectordistance'][0]
    z1                      = fid['/exchange/z1'][:]
    energy                  = fid['/exchange/energy'][0]
    shifts                  = fid['/exchange/shifts'][ids]
    shape                   = np.array(fid[f'/exchange/data0'].shape)

wavelength          = 1.24e-09 / energy
z2                  = focustodetectordistance - z1
magnifications      = focustodetectordistance / z1
norm_magnifications = magnifications / magnifications[0]
distances           = (z1 * z2) / focustodetectordistance * norm_magnifications**2
voxelsize           = detector_pixelsize / magnifications[0]
ndist = len(z1)
n     = shape[1]

print(f'{energy=}')
print(f'{z1=}')
print(f'{focustodetectordistance=}')
print(f'{detector_pixelsize=}')
print(f'{magnifications=}')
print(f'{voxelsize=}')
print(f'{distances=}')
```

### Convert encoder shifts to object-pixel coordinates

The motor encoder reports shifts in physical units for the *detector* plane.  Dividing by
`norm_magnifications[k]` maps each distance to the object plane at the highest magnification
(distance 0).  The (row, col) convention is also swapped here from (y, x) → (x, y).

```python
# shifts[ntheta, ndist, 2]: axis 2 is (y, x) in detector pixels.
# Convert to object-plane pixels and swap (y,x) → (row, col).
# norm_magnifications[ndist] broadcasts over [ntheta, ndist].
random_shifts = np.empty([ntheta, ndist, 2], dtype='float32')
random_shifts[..., 0] = shifts[..., 1] / norm_magnifications   # col (x) → row
random_shifts[..., 1] = shifts[..., 0] / norm_magnifications   # row (y) → col
```

#### RHAPP inter-plane alignment (from Peter's MATLAB pipeline)

`rhapp_python{str_z1_ids}.mat` contains the lateral offsets between the distance planes computed
by the RHAPP algorithm.  The axes are permuted and sign-flipped to match the Python (row, col)
convention used here.

```python
from holotomocupy.reader import load_octave_text_mat, load_shrink_from_mats

_rhapp_path = f'{path}/{pfile}_/rhapp.mat'
rhapp_raw = load_octave_text_mat(_rhapp_path, "rhapp")
rhapp_reordered = rhapp_raw.swapaxes(0, 2)[:ntheta]
rhapp_reordered -= rhapp_reordered[:,ref_dist:ref_dist+1]
avg_plane_zero = rhapp_reordered[:, 0].mean(axis=0)
rhapp_reordered -= avg_plane_zero[np.newaxis, np.newaxis, :]
print(f'avg_plane_zero  y={avg_plane_zero[0]:.4f} px   x={avg_plane_zero[1]:.4f} px')
rhapp_shifts = (-rhapp_reordered).astype("float32")

shrink_nd = load_shrink_from_mats(path, pfile, ndist, ntheta)
eff_magnifications = norm_magnifications / (1 + shrink_nd[0])
print(f'{shrink_nd.shape=}  {shrink_nd[0]=}')
```

### Motion shifts — slow drift of the reference plane

`correct_motion.txt` gives the stage position as a function of angle for a single reference
distance (index 2 here).  The **random_shifts at that distance are subtracted** so that only the
slow drift component remains, which is then replicated to all four distances.

```python
# Load file once, axis-flip (x,y)→(row,col) in one slice, convert to object plane.
# Subtract the encoder random_shifts at the reference distance so that
# only the slow drift component remains, then replicate to all distances.
raw_motion = np.loadtxt('/data2/vnikitin/atomium/20240924/AtomiumS2/AtomiumS2_HT_007nm_1_/correct_motion.txt')[:ntheta, ::-1].astype('float32')
motion_base = raw_motion / eff_magnifications[ref_dist] - random_shifts[:, ref_dist]
motion_shifts = np.tile(motion_base[:, np.newaxis], (1, ndist, 1))
```

### 3-D tomographic correction shifts

After an initial tomographic reconstruction, any remaining systematic misalignment of projections
shows up as blurring.  `correct_correct3D_{str_z1_ids}.txt` holds the per-angle correction
(same value applied to all distances, tiled with `np.tile`).

```python
raw_3d = np.loadtxt(f'/data2/vnikitin/atomium/20240924/AtomiumS2/AtomiumS2_HT_007nm_/correct_correct3D.txt')[:1800][:, ::-1].astype('float32')
correct3d_shifts = np.tile(raw_3d[:, np.newaxis], (1, ndist, 1))                # [ntheta, ndist, 2]
```

### Visualization

```python
plt.plot(motion_shifts[:,:,1])
plt.plot(motion_shifts[:,:,0])
plt.show()

plt.plot(correct3d_shifts[:,:,1])
plt.plot(correct3d_shifts[:,:,0])
plt.show()

plt.plot(rhapp_shifts[:,:,1])
plt.plot(rhapp_shifts[:,:,0])
plt.show()


plt.plot(random_shifts[:,:,1])
plt.plot(random_shifts[:,:,0])
plt.show()
```

#### Final combined shift = encoder + rhapp + motion + 3D correction

All four sources are additive; the result is saved as `/exchange/cshifts_final` (float32,
shape `[ntheta, ndist, 2]`) for use in Steps 4 and 5.

```python
shifts_final = rhapp_shifts + correct3d_shifts + motion_shifts + random_shifts
```

## Save to file

```python
with h5py.File(f'{path_out}/data.h5', 'a') as fid:
    if '/exchange/cshifts_final' in fid:
        del fid['/exchange/cshifts_final']
    fid.create_dataset('/exchange/cshifts_final', data=shifts_final.astype('float32'))
    if '/exchange/shrink' in fid:
        del fid['/exchange/shrink']
    fid.create_dataset('/exchange/shrink', data=shrink_nd)
```
