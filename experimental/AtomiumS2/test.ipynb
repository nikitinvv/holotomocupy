```python
def calc_ker(c,c1):
    n = 2048
    detector_pixelsize = 3.03751328912064e-6*np.sqrt(c1)
    energy = 33.35*c
    wavelength = 1.24e-09 / energy
    focustodetectordistance = 1.28

    sx0 = 1.286e-3
    z1 = 4.236e-3 - sx0
    z2 = focustodetectordistance - z1

    wavelength = 1.24e-09 / energy
    z2 = focustodetectordistance - z1
    magnification = focustodetectordistance / z1
    distance = (z1 * z2) / focustodetectordistance
    voxelsize = detector_pixelsize / magnification

    fx = cp.fft.fftfreq(2 * n,  d=voxelsize)
    fy = cp.fft.fftfreq(2 * n, d=voxelsize)
    fx, fy = cp.meshgrid(fx, fy)
    f2 = fx ** 2 + fy ** 2 

    fker= cp.exp(-1j * cp.pi * wavelength * distance * f2)
    ker = cp.fft.fftshift(cp.fft.ifft2(fker))
    return ker

c = 33.6/33.35#,17.23/17.1
v1 = calc_ker(1,c)
v2 = calc_ker(c,1)
print(cp.linalg.norm(v1-v2))
    
```
